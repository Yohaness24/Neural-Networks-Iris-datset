```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
print(type(iris)) 
print(type(iris.data), type(iris.target)) 
print(iris.data.shape, iris.target.shape)
print(iris.feature_names)
print(iris.target)
print (iris.target_names)
```

```python
features=iris.data
targets = iris.target
print(features, targets)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2)
X_train = X_train.T
X_test = X_test.T
print(X_train.shape)
y = np.zeros((3, 120))
for i in range(120):  
    y[y_train[i], i] = 1
print(y)
```

```python
#Activation functions
#sigmoid
def g1(t):
    return 1/(1 + np.exp(-t))
def g1_prime(t):
    return g1(t)*(1-g1(t))


# softmax
def g2(t):
    e = np.exp(t)
    return e / e.sum(axis=1, keepdims=True)
```

```python
# Parameters
n_x = 4
n_h = 30
n_y = 3
m = X_train.shape[1]
n = X_train.shape[0]
alpha = 0.01
print(n_x, n_h, n_y, m, n, alpha, X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
def NN(X_train, y_train, n_x, n_h, n_y, alpha, iterations):
    w_1 = np.random.randn(n_h, n_x)
    b_1 = np.zeros((n_h, 1))
    w_2 = np.random.randn(n_y, n_h)
    b_2 = np.zeros((n_y, 1))
    error_list = []
    for i in range(iterations):
            # Forward Prop
            z_1 = np.dot(w_1, X_train) + b_1
            a_1 = g1(z_1)
            z_2 = np.dot(w_2, a_1) + b_2
            a_2 = g2(z_2.T).T

            # Back Prop
            dz_2 = a_2 - y_train;
            dw_2 = np.dot(dz_2, a_1.T)/m
            db_2 = np.sum(dz_2, axis=1, keepdims=True)/m
            dz_1 = np.dot(w_2.T, dz_2) * g1_prime(z_1)
            dw_1 = np.dot(dz_1, X_train.T)/m
            db_1 = np.sum(dz_1, axis=1, keepdims=True)/m
            w_2 -= alpha * dw_2
            b_2 -= alpha * db_2
            w_1 -= alpha * dw_1
            b_1 -= alpha * db_1

            error = -1.0/m * np.sum(
                np.sum(y_train*np.log(a_2)) + 
                np.sum((1-y_train)*np.log(1-a_2))
            )
            error_list.append(error)
            
    return w_1, b_1, w_2, b_2, error_list

```

```python
# for Training
w_1, b_1, w_2, b_2, error_list = NN(X_train, y, n_x, n_h, n_y, alpha, 15000)
plt.plot(range(len(error_list)), error_list)
plt.show()
```

```python
z1 = np.dot(w_1, X_train) + b_1
a1 = g1(z1)
z2 = np.dot(w_2, a1) + b_2
a2 = g2(z2.T).T
print(a2)
pred = np.argmax(a2, axis=0)
print(pred)
score = accuracy_score(y_train, pred)
print(score)

cm = confusion_matrix(y_train, pred)
print(cm)
classification_report(y_train,pred)
print(classification_report(y_train,pred))
```

```python
#For Testing
w_1, b_1, w_2, b_2, error_list = NN(X_train, y, n_x, n_h, n_y, alpha, 15000)
plt.plot(range(len(error_list)), error_list)
plt.show()
z1 = np.dot(w_1, X_test) + b_1
a1 = g1(z1)
z2 = np.dot(w_2, a1) + b_2
a2 = g2(z2.T).T
print(a2)
pred = np.argmax(a2, axis=0)
print(pred)
score = accuracy_score(y_test, pred)
print(score)
```

```python
cm = confusion_matrix(y_test, pred)
print(cm)
classification_report(y_test,pred)
print(classification_report,pred)
```
